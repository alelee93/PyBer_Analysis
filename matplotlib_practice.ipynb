# 5.1.5 Create Bar Charts Using the MATLAB Approach

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot.
plt.bar(x_axis, y_axis)
```

```python
#Create the plot. 
plt.bar(x_axis, y_axis, color="green", label="Boston")

#Create labels for the x and y axes. 
plt.xlabel("date")
plt.ylabel("Fare($)")

#Create a title for the plot
plt.title("PyBerFare by Month")

#Add the legend
plt.legend()
```

```python
#Create the plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fare(#)")
plt.ylabel("date")
plt.title("PyFare by Month")
plt.legend()
```

```python
#Create the plot using the oo approach
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.legend()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBerFare by Month")
ax.invert_yaxis()
```

# 5.1.7 Create Scatter Plots and Bubble Charts

```python
#plt.plot(x_axis, y_axis, 'o')
plt.scatter(y_axis, x_axis, label="Chicago", c="red")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlabel("Fare($")
plt.ylabel("Date")
```

```python
# y_axis_larger = []
# for data in y_axis:
#     y_axis_larger.append(data*3)

# plt.scatter(x_axis, y_axis, s=y_axis_larger)

plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
#Create a scatter plot using OO interface
fig, ax = plt.subplots()
ax.scatter( y_axis, x_axis, s=y_axis, label="Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.legend()
ax.invert_yaxis()
```

# 5.1.8 Create Pie Charts


```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis, colors=colors, explode = explode_values, autopct='%.1f%%')


plt.show()
```
